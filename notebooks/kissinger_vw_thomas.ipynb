```python
# Set paths

# I use environment variables to set my base paths
DATA = os.environ['DATA']                                                                 
ME = os.environ['ME']                                                                     
# The following path agrees with kiss_to_text.py
MYDATA = os.path.join(DATA, ME, 'kiss')                                           
RAW = os.path.join(MYDATA, 'raw')                                                         
PROCESSED = os.path.join(MYDATA, 'processed')
IMAGES = os.path.join(PROCESSED, 'images')
CSV = os.path.join(PROCESSED, 'csv')
VW = os.path.join(PROCESSED, 'vw')

# You only need to set these paths below...any way you want...
metafile_path = os.path.join(RAW, 'meta', 'meta.csv')
bodyfiles_path = os.path.join(RAW, 'bodyfiles')
```

```python
#read in meta file
meta = pd.read_csv(metafile_path, sep='|').set_index('doc_id')
meta['date'] = meta['date'].apply(dateutil.parser.parse)

# Add Year-Month column
def clean_month(date):
    if pd.isnull(date):
        return date
    return date.strftime("%Y/%m")

meta['month'] = meta['date'].apply(clean_month)

meta.head()
```

```python
# Here I define some basic subclasses of the pos_tokenizer and
# streamer which cache the pos-tagging to speed of filtering by pos
# tags.

import os
from rosetta.text.text_processors import TokenizerPOSFilter
from rosetta.text.streamers import TextFileStreamer

import re
from rosetta.text import filefilter


class CachingTextFileStreamer(TextFileStreamer):
    def info_stream(self, paths=None, doc_id=None, limit=None):
        """
        Returns an iterator over paths yielding dictionaries with information
        about the file contained within.

        Parameters
        ----------
        paths : list of strings
        doc_id : list of strings or ints
        limit : Integer
            Use limit in place of self.limit.
        """
        if limit is None:
            limit = self.limit

        if doc_id is not None:
            paths = [self._doc_id_to_path[str(doc)] for doc in doc_id]
        elif paths is None:
            paths = self.paths

        for index, onepath in enumerate(paths):
            if index == limit:
                raise StopIteration

            with open(onepath, 'r') as f:
                text = f.read()
                doc_id = re.sub(
                    self.name_strip, '', filefilter.path_to_name
                    (onepath, strip_ext=False))
                stat_dict = self._file_stat(onepath)
                info_dict = {'text': text, 'cached_path': onepath,
                        'doc_id': doc_id}
                info_dict.update(stat_dict)
                if self.tokenizer:
                    info_dict['tokens'] = (
                        self.tokenizer.text_to_token_list(doc_id, text))

            yield info_dict

            
class CachingTokenizerPOSFilter(TokenizerPOSFilter):
    def __init__(self, *args, **kwargs):
        self.cache = {}
        super(CachingTokenizerPOSFilter, self).__init__(*args, **kwargs)

    def set_pos_types(self, pos_types):
        self.pos_types = set(pos_types)
    
    def text_to_token_list(self, doc_id, text):
        """
        Tokenize a list of text that (possibly) includes multiple sentences.
        """
        # sentences = [['I am Ian.'], ['Who are you?']]
        sentences = self.sent_tokenizer(text)
        # tokenized_sentences = [['I', 'am', 'Ian.'], ['Who', 'are', 'you?']]
        func = self.word_tokenizer.text_to_token_list
        tokenized_sentences = [func(sent) for sent in sentences]
        # tagged_sentences = [[('I', 'PRP'), ('am', 'VBP'), ...]]
        if self.cache.has_key(doc_id):
            tagged_sentences = self.cache[doc_id]
        else:
            tagged_sentences = [
                self.pos_tagger(sent) for sent in tokenized_sentences]
            self.cache[doc_id] = tagged_sentences

        # Returning a list of words that meet the filter criteria
        #token_list = sum(
        #    [self._sent_filter(sent) for sent in tagged_sentences], [])
        
        token_list = []
        for sent in tagged_sentences:
            token_list += self._sent_filter(sent)

        return token_list
```

```python
doc_tokens_file_path = os.path.join(VW, 'kiss_doc_tokens_NN_VB_JJ.vw')

my_tokenizer = CachingTokenizerPOSFilter(pos_types=['NN', 'VB', 'JJ'])
stream = CachingTextFileStreamer(text_base_path=bodyfiles_path,
                                 tokenizer=my_tokenizer)
stream.to_vw(doc_tokens_file_path, n_jobs=1)
```

```python
# This doesn't print stdout to the notebook. If you need to result, you should run
# the command directly in the shell. You can maybe use the subprocess module to get
# this to work nicer, but I couldn't get it to work. Something like:
#    proc = subprocess.Popen(CMD, stdout=subprocess.PIPE)
#    print proc.stdout.read()

num_topics = 5

CMD = "cd " + VW + ";"
CMD += "rm *.cache;"
CMD += "vw --lda " + str(num_topics) + " --cache_file doc_tokens.cache --passes 10 "
CMD += "    -p kiss_prediction_full.dat --readable_model kiss_topics_full.dat "
CMD += "    --bit_precision 16 " + doc_tokens_file_path

# Be careful! This opens a shell and runs previous command. It could overwrite something!
os.system(CMD)
```

##filter

```python
#Filter
sff_file_path = os.path.join(VW, 'sff_file.pkl')
sff = SFileFilter(VWFormatter())
sff.load_sfile(doc_tokens_file_path)

df = sff.to_frame()
df.head()
df.describe()

sff.filter_extremes(doc_freq_min=5, doc_fraction_max=0.8)
sff.compactify()
sff.save(sff_file_path)
```
