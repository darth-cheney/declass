```python
#get the DDRS meta file and clean it up a bit
meta = pd.read_csv(metafile_path, sep='\t', quotechar='|', na_values=['None'])
meta = meta.rename(columns={name: name.lower() for name in meta.columns})
meta = meta.set_index('id')
meta['written'] = pd.to_datetime(meta.written)
meta['released'] = pd.to_datetime(meta.released)
```

```python
#Add administration in there
admins = {                                                                                                        
 "Roosevelt" : (date(1933, 3, 4), date(1949, 1, 20)),                                                             
 "Truman": (date(1949, 1, 20), date(1953, 1, 20)),                                                                
 "Eisenhower" : (date(1953, 1, 20),  date(1961, 1, 20)),                                                          
 "Kennedy" : (date(1961, 1, 20), date(1963, 11, 22)),                                                             
 "Johnson" : (date(1963, 11, 22), date(1969, 1, 20)),                                                             
 "Nixon": (date(1969, 1, 20), date(1974, 8, 9)),                                                                  
 "Ford" : (date(1974, 8, 9), date(1977, 1, 20)),                                                                  
 "Carter": (date(1977, 1, 20), date(1981, 1, 20)),                                                                
 "Reagan" : (date(1981, 1, 20), date(1989, 1, 20)),                                                               
 "H. W. Bush" : (date(1989, 1, 20),  date(1993, 1, 20)),                                                          
 "Clinton" : (date(1993, 1, 20), date(2001, 1, 20)),                                                              
 "W. Bush" : (date(2001, 1, 20), date(2009, 1, 20)),                                                              
 "Obama": (date(2009, 1, 20), date(2017, 1, 20))                                                                  
}

func = lambda d: pd.Timestamp(datetime64(d))

admins = {name: (func(start), func(end)) for name, (start, end) in admins.iteritems()}

def get_admin(np_datetime, pres_dict):
    this_pres = 'unknown'
    for pres, (start, end) in pres_dict.iteritems():
        if start <= np_datetime <= end:
            this_pres = pres
            break
    
    return this_pres

meta['written_admin'] = meta.written.apply(lambda d: get_admin(d, admins))
meta['released_admin'] = meta.released.apply(lambda d: get_admin(d, admins))
```

```python
# The tokenizer turns strings of text such as "hi bye" into lists of tokens, e.g. ['hi', 'bye']
# You can add your own tokenizers to text_processors.  Give them the same API as TokenizerBasic.
tokenizer = text_processors.TokenizerBasic()
```

```python
sanitized_ids = []
unsanitized_ids = []
[sanitized_ids.append(index) if meta.loc[index]['sanitation'] else unsanitized_ids.append(index) for index in meta.index]
sanitized_paths = [os.path.join(CLEAN, str(docID) + '.clean'+'.txt') for docID in sanitized_ids]
unsanitized_paths = [os.path.join(CLEAN, str(docID) + '.clean'+'.txt') for docID in unsanitized_ids]
```

```python
###Tokenize
#Note: if you do not put a limit in the streamers the tokenization step can take a while (5min for the ddrs corpus on my macbook pro)

sanit_token_stream = text_processors.TokenStreamer(tokenizer=tokenizer, paths=sanitized_paths) 
unsanit_token_stream = text_processors.TokenStreamer(tokenizer=tokenizer, paths=unsanitized_paths) 
```

```python
t0 = time()
sanitized_tokens = []
for i, tokens in enumerate(sanit_token_stream):
    sanitized_tokens += tokens
    if i%10000==0: print 'done with: ', i

print 'done with sanitized'
unsanitized_tokens = []
for i, tokens in enumerate(unsanit_token_stream):
    unsanitized_tokens += tokens
    if i%10000==0: print 'done with: ', i
    
print 'sanitized tokens: ', len(sanitized_tokens)
print 'unsanitized tokens: ', len(unsanitized_tokens)
t1 = time()
print 'tokenization time: ', t1-t0, ' seconds'
```

```python
sanitized_FD = nltk.FreqDist(sanitized_tokens)
unsanitized_FD = nltk.FreqDist(unsanitized_tokens)
```

```python
def plot_top_freq(nltkFD, top, figsize=(15,10)):
    """
    plots freq dist from nltk.FreqDist object for top entries
    """
    plt.figure(figsize=figsize)
    values = nltkFD.values()[:top]
    values = np.array(values)/float(nltkFD.N())
    keys = nltkFD.keys()[:top]
    plt.plot(values)
    plt.xticks(range(len(values)), keys, rotation='vertical')
    
```
