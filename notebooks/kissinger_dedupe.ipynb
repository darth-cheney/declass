```python
import os
import pandas as pd
```

```python
# Set paths

# I use environment variables to set my base paths
DATA = os.environ['DATA']                                                                 
ME = os.environ['ME']                                                                     
MYDATA = os.path.join(DATA, ME, 'declass', 'kissinger-01')                                           
RAW = os.path.join(MYDATA, 'raw')                                                         
PROCESSED = os.path.join(MYDATA, 'processed')

# You only need to set these paths below...any way you want...
metafile_path = os.path.join(RAW, 'meta', 'meta.csv')
bodyfiles_path = os.path.join(RAW, 'bodyfiles')
```

```python
meta = pd.read_csv(metafile_path, sep='|').set_index('doc_id')
```

```python
meta.head()
```

```python
meta.loc[['0000C9F4','0000C9F2']]
```

```python
#remove rows with no time
meta = meta[meta.time!='None']
```

```python
meta.dropna()
```

```python
datetime_groups = meta.dropna().groupby(['date', 'time'])
```

```python
datetime_groups.groups
```

```python
group_keys = datetime_groups.groups.keys()
dupe_groups = []
dupe_groups_dict = {}
for key in group_keys:
    if len(datetime_groups.groups[key])>1:
        dupe_groups.append(datetime_groups.groups[key])
        dupe_groups_dict[str(key)] = str(datetime_groups.groups[key])

```

```python
dupe_groups_dict
```

```python
dupe_group_df = pd.DataFrame({'datetimes': dupe_groups_dict.keys(), 
                              "doc_ids": dupe_groups_dict.values()})
```

```python
dupe_group_df.to_csv('/Users/danielkrasner/Desktop/telcon_pairs.csv')
```

```python
def inspect_groups(groups):
    for g in groups:
        texts = []
        for doc_id in g:
            doc_id +='.txt'
            with open(os.path.join(bodyfiles_path, doc_id)) as f:
                  texts.append(f.read())
        yield texts
```

```python
ig = inspect_groups(dupe_groups)

```

```python
for group in ig.next():
    print "NEW DOC + \n"
    print group
```
