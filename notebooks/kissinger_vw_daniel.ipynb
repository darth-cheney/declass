```python
def parse_file_url(doc_id, results_json, master_url=('http://foia.state.gov/searchapp/'
            'DOCUMENTS/kissinger/')):
    """
    Parses the results_json object returned from the master_url and returns the all
    file urls. 
    """
    for doc_data in results_json:
        url_string = doc_data['pdfLink']
        if re.search(doc_id, url_string):
            doc = url_string.split('\\')[2]
            doc_id = doc.split('.')[0]
            doc_path = os.path.join(master_url, doc)
            return doc_path
```

```python
#clean up names a bit by throwing out Kissenger
def names_clean(name):
    if isinstance(name, str):
        name = name.upper()
    else:
        return name
    s = re.search(r'KISSINGER|SECRETARY', name)
    if s:
        name = re.sub(r"( AND |/SEC(RETARY)* |/|)KISSINGER[/]*", '', name)
        name = re.sub(r"(/| AND )THE SECRETARY", '', name)
    #some more formatting
    name = re.sub(r'GENERAL', 'G.', name)
    name = re.sub(r'SECRETARY', 'SECR.', name)
    name = re.sub(r' AND ', '/', name)
    name = re.sub(r'(MR|DR)[\.]*', '', name)
    name = re.sub(r'AMBASSADOR', 'AMBSDR', name)
    return name
    
```

```python
meta['names'] = meta['names'].apply(names_clean)
```
