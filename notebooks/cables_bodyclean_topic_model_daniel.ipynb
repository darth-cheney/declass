```python
### determine the optimal threshold
def opt_thresh(fpr, tpr, thresholds):
    """
    returns optimized roc threshold, i.e. the one which corresponds to
    min dist([fpr, tpr], [0,1])
    """
    roc = pd.DataFrame({'fpr': fpr, 'tpr': tpr})
    min_index = roc.apply(_roc_dist, axis=1).idxmin()
    return thresholds[min_index]
def _roc_dist(point):
    return dist(point, np.array([0, 1]))
```

```python
thresh = opt_thresh(fpr, tpr, thresholds)
```

```python
def get_class_stats(Y_test, Y_hat):
    """
    Returns basic classifier rate stats.
    """
    tpr = ((Y_hat==1)&(Y_test==1)).sum()/float((Y_hat==1).sum())
    fpr = ((Y_hat==1)&(Y_test==0)).sum()/float((Y_hat==1).sum())
    tnr = ((Y_hat==0)&(Y_test==0)).sum()/float((Y_hat==0).sum())
    fnr = ((Y_hat==0)&(Y_test==1)).sum()/float((Y_hat==0).sum())
    trues = (((Y_hat==1)&(Y_test==1)).sum()
             + ((Y_hat==0)&(Y_test==0)).sum())/float(Y_test.shape[0])
    return {'tpr': tpr, 'fpr': fpr, 'tnr': tnr, 'fnr': fnr, 'per_true': trues}
```
