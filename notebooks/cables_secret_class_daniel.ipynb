````python
#small cleaning for the "secret" meta variants
def clean_class(x):
    try:
        return re.sub(r'SECRET[\.\"\),S]', "SECRET", x)
    except TypeError:
        return x
        
````

````python
def get_result(clf, X):
    Y_score = clf.predict_proba(X.values)
    Yhat = np.argmax(Y_score, axis=1)
    conf = Y_score[np.arange(len(Y_score)), Yhat]
    #Yhat = le.inverse_transform(Yhat)
    
    result = pd.DataFrame({'Yhat': Yhat, 'conf': conf}, index=X.index)
    #result.Yhat = result.Yhat.apply(lambda x: 'SECRET' if 1 else 'NOT-SECRET')
    
    return result
````
